```python
import pandas as pd
df = pd.read_csv("supply_chain_extended_data.csv")

# Group by Warehouse and Location, and calculate average times
delivery_stats = df.groupby(['Warehouse_ID', 'Location'])[
    ['Lead_Time_Days', 'Shipping_Time_Days', 'Order_Processing_Time']
].mean().reset_index()

# Total delay
delivery_stats['Total_Delay'] = (
    delivery_stats['Lead_Time_Days'] +
    delivery_stats['Shipping_Time_Days'] +
    delivery_stats['Order_Processing_Time']
)

# Top 10 delayed warehouses
top_delayed_warehouses = delivery_stats.sort_values(by='Total_Delay', ascending=False).head(10)
print(top_delayed_warehouses)


```

```python
# Stock Gap
df['Stock_Gap'] = df['Current_Stock'] - df['Demand_Forecast']

# Top 5 understocked (shortage risk)
understocked = df.sort_values(by='Stock_Gap').head(5)

# Top 5 overstocked (excess inventory)
overstocked = df.sort_values(by='Stock_Gap', ascending=False).head(5)

print("Understocked Warehouses:\n", understocked[['Warehouse_ID', 'Location', 'Current_Stock', 'Demand_Forecast', 'Stock_Gap']])
print("\nOverstocked Warehouses:\n", overstocked[['Warehouse_ID', 'Location', 'Current_Stock', 'Demand_Forecast', 'Stock_Gap']])
```

```python
# Forecast error
df['Forecast_Error'] = df['Demand_Forecast'] - df['Monthly_Sales']

# Top 5 over-forecasted (high positive error)
over_forecasted = df.sort_values(by='Forecast_Error', ascending=False).head(5)

# Top 5 under-forecasted (high negative error)
under_forecasted = df.sort_values(by='Forecast_Error').head(5)

print("Over-Forecasted:\n", over_forecasted[['Warehouse_ID', 'Location', 'Demand_Forecast', 'Monthly_Sales', 'Forecast_Error']])
print("\nUnder-Forecasted:\n", under_forecasted[['Warehouse_ID', 'Location', 'Demand_Forecast', 'Monthly_Sales', 'Forecast_Error']])
```

```python
# Profitability (approximation)
df['Profitability'] = df['Monthly_Sales'] - (
    df['Operational_Cost'] + df['Storage_Cost'] + df['Transportation_Cost']
)

# Top 5 most profitable
top_profitable = df.sort_values(by='Profitability', ascending=False).head(5)

# Top 5 least profitable
least_profitable = df.sort_values(by='Profitability').head(5)

print("Most Profitable:\n", top_profitable[['Warehouse_ID', 'Location', 'Monthly_Sales', 'Operational_Cost', 'Storage_Cost', 'Transportation_Cost', 'Profitability']])
print("\nLeast Profitable:\n", least_profitable[['Warehouse_ID', 'Location', 'Monthly_Sales', 'Operational_Cost', 'Storage_Cost', 'Transportation_Cost', 'Profitability']])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df['Total_Delay'] = df['Lead_Time_Days'] + df['Shipping_Time_Days'] + df['Order_Processing_Time']
# Top 10 delayed warehouses
top_delayed = df.groupby(['Warehouse_ID', 'Location'])['Total_Delay'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_delayed.plot(kind='barh', color='red')
plt.title("Top 10 Warehouses by Delivery Delay")
plt.xlabel("Average Total Delay (Days)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
df['Stock_Gap'] = df['Current_Stock'] - df['Demand_Forecast']
# Overstock vs Understock
plt.figure(figsize=(10, 6))
sns.histplot(df['Stock_Gap'], bins=30, kde=True, color='blue')
plt.axvline(0, color='red', linestyle='--')
plt.title("Inventory Gap Distribution")
plt.xlabel("Stock Gap (Current Stock - Forecast)")
plt.show()
```

```python
df['Forecast_Error'] = df['Demand_Forecast'] - df['Monthly_Sales']
# Scatter plot of forecast vs sales
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Demand_Forecast', y='Monthly_Sales', hue='Product_Category')
plt.plot([df['Demand_Forecast'].min(), df['Demand_Forecast'].max()],
         [df['Demand_Forecast'].min(), df['Demand_Forecast'].max()],
         color='gray', linestyle='--')  # Perfect forecast line
plt.title("Forecast vs Actual Sales")
plt.show()
```

```python
df['Profitability'] = df['Monthly_Sales'] - (
    df['Operational_Cost'] + df['Storage_Cost'] + df['Transportation_Cost']
)
# Top & bottom 5 profitability
top5_profit = df.sort_values(by='Profitability', ascending=False).head(5)
bottom5_profit = df.sort_values(by='Profitability').head(5)

# Combine and plot
plt.figure(figsize=(10, 6))
sns.barplot(data=pd.concat([top5_profit, bottom5_profit]), 
            x='Warehouse_ID', y='Profitability', hue='Location')
plt.title("Top and Bottom 5 Warehouses by Profitability")
plt.xticks(rotation=45)
plt.show()
```

```python
# Map-like bubble chart (simulated with scatter)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Location', y='Total_Delay', size='Total_Delay', hue='Location', legend=False, sizes=(50, 500))
plt.title("Delay Distribution by Warehouse Location")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Stacked bar: Current Stock vs Forecast
inv_sample = df.groupby('Warehouse_ID')[['Current_Stock', 'Demand_Forecast']].mean().head(10)
inv_sample.plot(kind='bar', stacked=False, figsize=(10, 6), color=['skyblue', 'orange'])
plt.title("Stock vs Forecast by Warehouse (Sample)")
plt.ylabel("Units")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Forecast Error distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Forecast_Error'], bins=30, kde=True, color='purple')
plt.title("Forecast Error Distribution")
plt.xlabel("Forecast Error")
plt.tight_layout()
plt.show()
```

```python
# Cost Breakdown
cost_components = df[['Operational_Cost', 'Storage_Cost', 'Transportation_Cost']].sum()
plt.figure(figsize=(8, 6))
cost_components.plot(kind='bar', color=['green', 'blue', 'orange'])
plt.title("Total Cost Breakdown")
plt.ylabel("Cost")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
